Labelling image from dataset for Yolo4
Yolo4 expect format <object-class> <x_center> <y_center> <width> <height>
labels in this dataset have format x_from	y_from	width	height	sign_class

```python
import os
import pandas as pd
import shutil
```

```python
allFileNames = os.listdir('images')
```

```python
data_df = pd.read_csv('full-gt.csv')
data_df.head()
```

```python
from PIL import Image

im = Image.open('images/' + allFileNames[1000])
width, height = im.size
print(width)
print(height)
```

```python
d = {}
with open('classes_map.txt') as f:
    for line in f:
        (key, val) = line.split()
        d[key] = val
```

```python
d
```

```python
d.get('1_15')
```

```python
classes = []
more_than_one = []

for name in allFileNames:
    file_name, file_extension = os.path.splitext(name)
    
    im = Image.open('images/' + name)
    width, height = im.size
    

    file_name = file_name + '.txt'

    part_df = data_df[data_df['filename'] == name]

    for index, row in part_df.iterrows():
        x_center = float((row['x_from'] + int(row['width'] / 2)) / width)
        y_center = float((row['y_from'] + int(row['height'] / 2)) / height)
        
        w = float(row['width'] / width)
        h = float(row['height'] / height)
        
        txt_row = d.get(row['sign_class']) + ' ' + str(x_center) + ' ' + str(y_center) + ' ' + str(w) + ' ' + str(h)
        
        if x_center > 1 or y_center > 1:
            more_than_one.append(file_name)

        #classes.append(row['sign_class'])
        
        with open('txt/' + file_name, 'a') as out:
            out.write(txt_row + '\n')
```
